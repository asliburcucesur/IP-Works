```python
import pandas as pd
df = pd.read_csv("country_vaccination_stats.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df["daily_vaccinations"] = df.groupby("country")["daily_vaccinations"].transform(lambda x: x.fillna(x.min()))
df
```

```python
df.isnull().sum()
```

```python
df["daily_vaccinations"]=df["daily_vaccinations"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.loc[df["country"]=="Kuwait"]
```

```python
median_df= df.groupby("country")[["daily_vaccinations"]].agg(lambda x: x.astype(int).median())
```
